```python
#Remove duplicated rows
def drop_duplicated(df):
    duplicated_titles = df[df['title'].duplicated(keep=False)]
    new_df = df.copy()
    for title in duplicated_titles['title'].unique():
        filter_duplicated_titles = duplicated_titles.loc[duplicated_titles['title']==title]
        # Armazenando os índices das linhas em uma lista
        index = filter_duplicated_titles.index.tolist()
        print(index)
        for id in index:
            id_max_pages = filter_duplicated_titles['num_pages'].idxmax()
            if id != id_max_pages:
                new_df.drop(id, inplace=True)   
                print('ID removed:', id)
        print('----------------------------------')
    return new_df
```

# <font color='darkblue' style='font-size: 20;'>Building the model</font>
<hr style = 'border: 1px solid darkblue;'>

```python
#Drop unecessary columns
new_df.drop(columns=['bookID', 'isbn', 'isbn13', 'language_code', 'ratings_count', 'text_reviews_count', 'publication_date'], axis=1, inplace=True)

# Resetar o índice do DataFrame
new_df.reset_index(drop=True, inplace=True)

#Create a new book_id
new_df['book_id'] = new_df.index
```

### <font color='darkblue' style='font-size: 20;'>Cosine similarity - text columns</font> 

```python
#Create a list of text columns
columns = ['title', 'authors', 'publisher']

#Create a function to combine the important columns/features
def combine_features(data):
    features = []
    for i in range(0, data.shape[0]):
        features.append(data['title'][i] + ' ' + data['authors'][i] + ' ' + data['publisher'][i])
    return features

#Create a column to store the combined features
new_df['combined_features'] = combine_features(new_df)
new_df.head()
```

```python
from sklearn.metrics.pairwise import cosine_similarity
from sklearn.feature_extraction.text import CountVectorizer
```

```python
#Convert the text from the new column to a matrix of word counts
cm = CountVectorizer().fit_transform(new_df['combined_features'])
```

```python
#Get the cosine similarity matrix from the count matrix
cs = cosine_similarity(cm)
#print the scores
print(cs)
```

### <font color='darkblue' style='font-size: 20;'>Cosine similarity - numeric columns</font> 

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
#Normalize the numeric data
scaler = MinMaxScaler()
numeric_data = new_df[['average_rating', 'num_pages']]
normalized_data = scaler.fit_transform(numeric_data)
```

```python
#Cosine similarity
numeric_similarity = cosine_similarity(normalized_data)
numeric_similarity
```

```python
#Combination of the consine similarities
similarity = 0.7 * cs + 0.3 * numeric_similarity
```

```python
#Get the title of the book the reader likes
title = new_df['title'][0]
title
```

```python
#Find the book id of the book that the user like
book_id = new_df[new_df.title == title]['book_id'].values[0]
book_id
```

```python
#Create a list of tuples in the form (index, similarity score)
scores = list(enumerate(similarity[book_id]))
print(scores)
```

```python
#Sort the list of similar books in descending order
sorted_scores = sorted(scores, key=lambda x:x[1], reverse=True)
sorted_scores = sorted_scores[1:]
sorted_scores
```

```python
#Create a loop to print 5 recommended books
j = 0
print('The most recommended books to '+title+' are:\n')
for item in sorted_scores:
    book_title = new_df[new_df.book_id == item[0]]['title'].values[0]
    print(j+1, book_title)
    j = j+1
    if j>=5:
        break
```

### <font color='darkblue' style='font-size: 20;'>Final model</font> 

```python
#Create a function
def book_recommender(title):
    
    #Find the book id of the book that the user like
    book_id = new_df[new_df.title == title]['book_id'].values[0]
    
    #Create a list of tuples in the form (index, similarity score)
    scores = list(enumerate(similarity[book_id]))
    
    #Sort the list of similar books in descending order
    sorted_scores = sorted(scores, key=lambda x:x[1], reverse=True)
    sorted_scores = sorted_scores[1:]
    
    print('The most recommended books to '+title+' are:\n')
    for item in sorted_scores[:5]:
        book_title = new_df[new_df.book_id == item[0]]['title'].values[0]
        print(book_title)
```
