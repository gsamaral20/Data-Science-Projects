#### <font color='darkblue' style='font-size: 20;'>Hypothesis Test</font> 

**H<sup>null</sup>**
>**The tip rate distribution is the same in both groups**

**H<sup>alt</sup>**
>**The tip rate distribution is not the same in the two groups**

```python
from scipy.stats import ranksums
```

```python
with_dessert = df.query("dessert == 'Yes'").percentage
without_dessert = df.query("dessert == 'No'").percentage
```

```python
r = ranksums(with_dessert, without_dessert)
```

```python
print("The value of p-value is: {}".format(r.pvalue.round(4)))
```

The p-value represents the probability of that sample having happened within the population. If the chance is small, usually p-value < 0.05, it represents that a very rare event happened, so we choose to discard the null hypothesis, and say that it may not be true.

That is, it represents that we do not have enough evidence to accept the alternative hypothesis.

#### <font color='darkblue' style='font-size: 20;'>Hypothesis Test</font> 

**H<sup>null</sup>**
>**Distribution of account value is the same on Saturday and Sunday**

**H<sup>alt</sup>**
>**The distribution of the account value is not equal on Saturday and Sunday**

```python
sunday_bill = df.query("day == 'Sun'").total_bill
saturday_bill = df.query("day == 'Sat'").total_bill

r2 = ranksums(sunday_bill, saturday_bill)
print('The value of p-value is {}'.format(r2.pvalue.round(4)))
```

The result of pvalue will be 0.35, that is, we will only accept the alternative hypothesis if this number is smaller than 0.05; since it is larger, only our null hypothesis is valid. This means that the distribution of the bill value is equal on Saturday and Sunday.

Therefore, even though Saturday is more crowded and Sunday has the highest average tipping, they are not different with respect to the value of the bill.

#### <font color='darkblue' style='font-size: 20;'>Hypothesis Test</font> 

**H<sup>null</sup>**
>**The distribution of the bill is the same at lunch and dinner**

**H<sup>alt</sup>**
>**The distribution of the bill is not the same at lunch and dinner**

```python
df.groupby(['time']).mean().round(3)[['total_bill', 'tip', 'percentage']]
```

```python
r3 = ranksums(dinner, lunch)
print('The value of p-value is {}'.format(r3.pvalue.round(4)))
```

The system returns a pvalue less than 0.05, which will mean that the distribution is different in the two periods of the day and that the null hypothesis can be rejected.

```python
perc_lunch = df.query("time == 'Lunch'").percentage
perc_dinner = df.query("time == 'Dinner'").percentage
```

```python
r4 = ranksums(perc_dinner, perc_lunch)
print('The value of p-value is {}'.format(r4.pvalue.round(4)))
```

The pvalue is greater than 0.05, which will mean that we have similar groups. That is, the tipping fee distribution is indeed equal at lunch and dinner.
